```python
import os 
base_dir = "/Users/eyubogln/Downloads/hapi_raw/"

# load all the text files at path {base_dir}/{dataset}/ReadMe.txt
# and return a dictionary of {dataset}: {readme}
readme_dict = {}
for dataset in os.listdir(base_dir):
    readme_path = os.path.join(base_dir, dataset, "ReadMe.txt")
    if not os.path.exists(readme_path):
        continue
    with open(readme_path, "r") as f:
        readme = f.read()
    readme_dict[dataset.lower()] = readme

# write the dictionary to json at /Users/eyubogln/code/hapi/hapi/data.json
import json 
with open("/Users/eyubogln/code/hapi/hapi/data.json", "w") as f:
    json.dump(readme_dict, f, indent=4)
```
